```python
# step 1: import library
import pandas as pd
```

```python
# step 2: import data
Flipkart = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Flipkart%20Headphones.csv')
```

```python
Flipkart.head()
```

```python
Flipkart.info()
```

```python
Flipkart.describe()
```

```python
# step 3: Define target (y) and features (X)
```

```python
Flipkart.columns
```

```python
y = Flipkart['Discount']
```

```python
X = Flipkart.drop(['Model', 'Company', 'Color', 'Type', 'Average Rating',
       'Number of Ratings', 'Selling Price', 'Maximum Retail Price'], axis = 1)
```

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=2529)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Step 5: select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Step 6: train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Step 7: predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Step 8 : model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
